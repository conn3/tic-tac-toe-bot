```python

#display the board in tictactoe style 
def printBoard(table):
    print()
    for x in range(len(table)):
        for y in range(len(table[x])):
            print(" |",table[x][y],end="")
        print(" |")
        if(x!=2):
            print("----------------")
#check for tie condition to end game 
def checkTie(board):
    for i in range(3):
        for j in range(3):
            if(board[i][j]==' '):
                return False
    return True
#function to check if game is over    
def checkWinner(table):
    isFull = True
    for i in range(3):
        #check rows
        if(table[i][0] == "x" and table[i][1] == "x" and table[i][2] == "x"):
            return 1
        elif(table[i][0] == "o" and table[i][1] == "o" and table[i][2] == "o"):
            return -1
        #check cols
        if(table[0][i] == "x" and table[1][i] == "x" and table[2][i] == "x"):
            return 1
        elif(table[0][i] == "o" and table[1][i] == "o" and table[2][i] == "o"):
            return -1
        #check diagonal
        if(table[0][0] == "x" and table[1][1] == "x" and table[2][2] == "x"):
            return 1
        elif(table[0][0] == "o" and table[1][1] == "o" and table[2][2] == "o"):
            return -1
        if(table[0][2] == "x" and table[1][1] == "x" and table[2][0] == "x"):
            return 1
        elif(table[0][2] == "o" and table[1][1] == "o" and table[2][0] == "o"):
            return -1
        for j in table[i]:
            if(j == ' '):
                isFull = False
    if(isFull == True):
        return 0
    elif(isFull == False):
        return None

def checkBoard(table):
    for i in range(3):
        #check rows
        if(table[i][0] == "x" and table[i][1] == "x" and table[i][2] == "x"):
            printBoard(table)
            print("x wins!")
            return 1
        elif(table[i][0] == "o" and table[i][1] == "o" and table[i][2] == "o"):
            printBoard(table)
            print("o wins!")
            return -1
        #check cols
        if(table[0][i] == "x" and table[1][i] == "x" and table[2][i] == "x"):
            printBoard(table)
            print("x wins!")
            return 1
        elif(table[0][i] == "o" and table[1][i] == "o" and table[2][i] == "o"):
            printBoard(table)
            print("o wins!")
            return -1
        #check diagonal
        if(table[0][0] == "x" and table[1][1] == "x" and table[2][2] == "x"):
            printBoard(table)
            print("x wins!")
            return 1
        elif(table[0][0] == "o" and table[1][1] == "o" and table[2][2] == "o"):
            printBoard(table)
            print("o wins!")
            return -1
        if(table[0][2] == "x" and table[1][1] == "x" and table[2][0] == "x"):
            printBoard(table)
            print("x wins!")
            return 1
        elif(table[0][2] == "o" and table[1][1] == "o" and table[2][0] == "o"):
            printBoard(table)
            print("o wins!")
            return -1
    return 0

#play function
def play():
    #play the ai
    board = [[' ',' ', ' '], [' ', ' ', ' '], [' ', ' ', ' ']]
    move = 0
    while(move < 9 or checkBoard(board) != 0):
      
        #ai move
        ai = bMove(board)
        print("--------------------------------------")
        print("Turn #",move+1)
        print("AI places on",ai) 
        board[ai[0]][ai[1]] = "x"
        if(move >= 9 or checkBoard(board) != 0):
            printBoard(board)
            break
        printBoard(board)
        if(checkTie(board)):
            print("Tie!")
            break
        #loop until valid input
        while(True):
            print()
            userInput = int(input("Enter 1-9 to place an 'o': ")) 
            userInput=userInput-1
            playerX = userInput//3
            playerY = userInput%3
            if(board[playerX][playerY]==' '):
                board[playerX][playerY] = "o"
                break
            print("invalid move position already occupied"," ")
        printBoard(board)
        move += 1
        if(checkBoard(board) != 0):
            break

#best move for ai
def bMove(table):
    bestMove = [0, 0]
    bestScore = -9999999
    for i in range(len(table)):
        for j in range(len(table[i])):
            if(table[i][j] == ' '):
                table[i][j] = "x"
                score = minimax(table, 0, False)
                table[i][j] = ' '
                if(score > bestScore):
                    bestScore = score
                    bestMove = [i, j]
                    
    return bestMove
            
#minimax algorithm
def minimax(table, depth, isMaxing):
    result = checkWinner(table)
   
    if(result != None):
        score = result
        return score
    
    if(isMaxing):
        bestScore = -9999999
        for i in range(len(table)):
            for j in range(len(table[i])):
                if(table[i][j] == ' '):
                    table[i][j] = "x"
                    score = minimax(table, depth+1, False)
                    table[i][j] = ' '
                    if(score > bestScore):
                        bestScore = score
        return bestScore
    else:
        bestScore = 9999999
        for i in range(len(table)):
            for j in range(len(table[i])):
                if(table[i][j] == ' '):
                    table[i][j] = "o"
                    score = minimax(table, depth+1, True)
                    table[i][j] = ' '
                    if(score < bestScore):
                        bestScore = score
        return bestScore
play()
            
            
```
